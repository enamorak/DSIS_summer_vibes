Этот график поможет нам увидеть, есть ли какая-либо зависимость между временем между первой и последней транзакции и количеством отправленных транзакций. Заметим, что чем больше разница между временем первой и последней транзакции, тем меньше отправленных транзакций.

Эта связь между разницей во времени между первой и последней транзакций и количеством отправленных транзакций связана с концепцией утечек Ethereum, известных как "фронтраннинг" или "мемпул-арбитраж". 

Утечки Ethereum происходят, когда злоумышленники получают информацию о предстоящих транзакциях, еще не включенных в блокчейн, и предпринимают меры для выполнения своих собственных транзакций, которые могут извлечь выгоду из этой информации.

Когда пользователь отправляет транзакцию на Ethereum, она сначала попадает в мемпул, который является временным хранилищем для ожидающих обработки транзакций. Затем майнеры выбирают транзакции из мемпула и включают их в новые блоки, добавляя их в блокчейн.

Злоумышленники, использующие фронтраннинг, могут наблюдать мемпул и видеть, какие транзакции ожидаются. Если они узнают о предстоящей транзакции, которая может повлиять на рынок или предоставить им выгоду, они могут быстро отправить свою собственную транзакцию с более высокой комиссией, чтобы ее включили в блок раньше, чем оригинальную транзакцию пользователя.

Таким образом, злоумышленники могут использовать информацию о разнице во времени между первой и последней транзакций для ускорения своих собственных транзакций и получения преимущества перед другими участниками сети. Они смогут отправить свои транзакции в последние моменты, непосредственно перед исполнением других транзакций, чтобы уменьшить возможность конкуренции и повысить вероятность включения своих транзакций в следующий блок.

В ответ на эти утечки Ethereum, сообщество разработчиков и исследователей предпринимает шаги для улучшения протокола и снижения возможности фронтраннинга. Например, были предложены различные механизмы, такие как введение "газовых потоков" (поскольку каждая операция требует определенное количество газа, злоумышленники не могут перегрузить сеть, отправляя большое количество недорогих транзакций или выполняя сложные операции без оплаты достаточной стоимости в газе) и оптимизация работы мемпула, для борьбы с этими утечками и повышения защиты и справедливости в сети Ethereum.

```python
# Построение графика для среднего времени между отправленными транзакциями
plt.hist(df['Avg min between sent tnx'])
plt.xlabel('Average min between sent tnx')
plt.ylabel('Count')
plt.title('Distribution of Average min between sent tnx')
plt.show()
```

Заметим, что чем меньше среднее временя между отправленными транзакциями, тем больше транзакций.

Это связано с тем, что если между отправкой транзакций проходит меньше времени, то пользователь может отправить больше транзакций в течение определенного временного интервала. 

Например, если среднее время между отправкой транзакций составляет 1 минуту, то пользователь может отправить до 60 транзакций в час. Однако, если среднее время между отправкой транзакций увеличивается до 5 минут, то пользователь сможет отправить только 12 транзакций в час.

Данная особенность отправки используется при выполнении массовых платежей.

```python
# Построение графика для среднего времени между полученными транзакциями
plt.hist(df['Avg min between received tnx'])
plt.xlabel('Average min between received tnx')
plt.ylabel('Count')
plt.title('Distribution of Average min between received tnx')
plt.show()
```

Заметим, что чем меньше среднее временя между полученными транзакциями, тем больше транзакций. 

Если некоторые пользователи регулярно отправляют транзакции с одного адреса на другой, это может позволить третьим лицам сделать выводы о связях между этими адресами и их владельцами. Это нарушает приватность пользователей.

Если среднее время между транзакциями мало, то это может указывать на то, что множество операций выполняется в короткий промежуток времени. Если в этих операциях есть газовые утечки, то это может привести к ненужным затратам эфира и утечкам средств.

```python
import plotly.express as px
# Создание графика типа наиболее принимаемого токена ERC20
fig = px.pie(data_frame=df, names=' ERC20_most_rec_token_type', title='ERC20 Most Received Token Types')
fig.show()
```

Если не обращать внимания на нулевые значения в датасете, заметим, что наиболее часто получаемые типы токенов - OmiseGo (8.87%), NOTSAFU (7.92%), DATAcoin (3.64%). Проанализируем этот результат.

OmiseGo был запущен на базе Ethereum в 2017 году и является одним из наиболее популярных токенов на этой платформе. Утечки Ethereum могут быть связаны с его популярностью и широким использованием, что привлекает внимание злоумышленников.

NOTSAFU предоставляет возможность владельцам утраченных средств получить некоторую компенсацию. Участие в таких утечках может привести к получению этого типа токена.

DATAcoin используется для "51% атаки" криптовалютной сети - злоумышленник получает контроль над большей частью вычислительной мощности сети. 

```python
uniq_sent_tokens = df[' ERC20 uniq sent token name']
uniq_rec_tokens = df[' ERC20 uniq rec token name']

plt.figure(figsize=(8, 8))
plt.pie([uniq_sent_tokens.sum(), uniq_rec_tokens.sum()], labels=['Unique Sent Tokens', 'Unique Received Tokens'], autopct='%1.1f%%')
plt.title('Distribution of Unique Sent and Received Tokens')
plt.show()
```

Круговая диаграмма распределения уникально отправленных и полученных токенов показывает, что уникальных полученных токенов в 3.5 раза больше, чем уникальных отправленных токенов. Это связано с тем, что утечки Ethereum могут происходить с одного адреса, но токены могут быть получены различными адресами в сети Ethereum.

Когда происходит утечка, токены все равно могут быть получены другим адресом в сети Ethereum. Это может быть адрес, созданный злоумышленником, который получает утечку Ethereum, или адрес, принадлежащий другому пользователю, который случайно получает утечку Ethereum.

Этот график позволяет увидеть потенциальные выбросы или общие характеристики среди мошеннических адресов. Заметим, что общая характеристика - отправка/получение максимально возможного количества токенов за короткое время. Почему так происходит?

Каждая операция требует определенного количества газа, которое является единицей измерения для расчета стоимости выполнения операции. Если операциям требуется слишком много газа, они могут быть не включены в блок из-за его переполнения. Поэтому отправка/получение большого количества токенов за короткое время может привести к утечкам, если операции не могут быть выполнены из-за ограничений газа.

При отправке/получении большого количества токенов за короткое время может возникнуть сетевая задержка. В результате этого могут возникнуть утечки, так как отправка/получение токенов не будет происходить в ожидаемые сроки.

Если контракт не был должным образом проверен и защищен, злоумышленники могут использовать его для несанкционированного доступа к токенам или выполнения других вредоносных действий, как только стало известно об уязвимости.
